# marble game
probability and combination of marble game. 3 colors marble, each 8 pieces. Take 12. See whether the combination of color you got is according to the winning combination

![Example Image](D:\martin\python_code\just4fun\images\marble_win.jpg)

```python
from itertools import combinations

# Sample list
n=8
my_list = n*['R']+n*['G']+n*['B']

# Generate all possible combinations of length 2
result = list(combinations(my_list, 12))

# Display the result
#print(result)
```

```python
len(result)
```

```python
from collections import Counter

combination={}

for i in result:
    element_counts = Counter(i)
    key=''.join(map(str, sorted(list(element_counts.values()))))
    
    if key in combination:
        combination[key] += 1
    else:
        combination[key] = 1
```

```python
# Sort the dictionary by values in descending order
sorted_dict = dict(sorted(combination.items(), key=lambda item: item[1], reverse=True))

sorted_dict
```

```python
import matplotlib.pyplot as plt

# Extract labels and values from the dictionary
labels = list(sorted_dict.keys())
values = list(sorted_dict.values())
```

```python
# Calculate the total sum for percentage calculation
total_sum = sum(values)

# Calculate percentages
percentages = [(value / total_sum) * 100 for value in values]

# Plotting the bar chart
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.bar(labels, percentages)

# Adding percentage values above each bar
for bar, percentage in zip(bars, percentages):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, height + 0.5, f'{percentage:.2f}%', ha='center', va='bottom')

plt.xlabel('Categories')
plt.ylabel('Percentage')
plt.title('Distribution of Values')
plt.show()
```

```python
# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(values, labels=None, autopct='%1.1f%%', startangle=140, labeldistance=1.1)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Adding legend outside the pie chart
plt.legend(labels, loc='center left', bbox_to_anchor=(1, 0.5))

plt.title('Distribution of Values')
plt.show()
```
